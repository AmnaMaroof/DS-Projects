### np.random.randint
(low, high, size): Generates random integers from low (inclusive) to high (exclusive) with the specified size (number of integers).
### np.column_stack(arrays): 
 This combines the individual arrays (columns) into a single 2D array (matrix). Each column corresponds to one of the attributes and each row corresponds to one data point.
Stacks the input arrays as columns into a single 2D array. This results in a 2D array where each row represents a data point with all attributes.

```python
np.random.seed(42)
random_size = 10

student_id=np.random.randint(1, 20, random_size)
test_score=np.random.randint(25, 100, random_size)
writing_skill=np.random.randint(25, 100, random_size)
reading_skill=np.random.randint(25, 100, random_size)
study_hours=np.random.randint(2, 10, random_size)
attendance_percentage=np.random.randint(75, 100, random_size)

# Combine the arrays into a dataset
Data_set_2= np.column_stack((student_id, test_score, writing_skill, reading_skill, study_hours, attendance_percentage))

# Convert to DataFrame
Data_set_2 = pd.DataFrame(Data_set_2, columns=['student_id', 'test_score', 'writing_skill', 'reading_skill', 'study_hours', 'attendance_percentage'])

print(Data_set_2)
```
